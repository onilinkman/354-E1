**3. Del dataset elegido realice la imputaci√≥n por columnas sin scikit-learn.**

```python
import pandas as pd
dat = pd.read_csv("./bank.csv")
dat.head(10)
```

```python
import pandas as pd
pd.options.mode.chained_assignment = None
imp = dat.iloc[:, [0, 5, 9, 11, 12, 13, 14]]
# Imputar valores faltantes en las columnas seleccionadas con la mediana de cada columna
for column in imp.columns:
    median = imp[column].median()
    imp[column].fillna(median, inplace=True)
imp.head(10)
```

En el contexto del conjunto de datos "bank marketing", la imputaci√≥n puede ser necesaria debido a la presencia de valores faltantes en las caracter√≠sticas o atributos de los registros. Estos valores faltantes pueden surgir por diversas razones, como errores en la recopilaci√≥n de datos, informaci√≥n incompleta o simplemente porque algunos registros no proporcionan ciertos detalles.

La imputaci√≥n se utiliza para llenar estos valores faltantes de manera que el conjunto de datos sea m√°s completo y adecuado para su an√°lisis.
