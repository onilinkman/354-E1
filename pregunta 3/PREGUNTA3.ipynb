### SimpleImputer
#### Uso: Rellenar valores faltantes en el dataset.
#### Razon de Uso: Los valores faltantes pueden llevar a errores en el analisis y en la construccion de modelos. Imputar estos valores garantiza que todos los datos sean utilizables.

```python
from sklearn.impute import SimpleImputer
#Imputer valores nulos con la media para caracteristicas numericas
imputer=SimpleImputer(strategy='mean')
data[['Total','gross income']]=imputer.fit_transform(data[['Total','gross income']])
```

```python
data.head()
```

### OneHotEncoder(Codificacion de Variables Categoricas)
#### Uso: Convertir variables categóricas en variables dummy/indicadoras.
#### Razón de Uso: Los algoritmos de aprendizaje automático requieren entradas numéricas. La codificación one-hot transforma categorías en un formato que los modelos pueden entender.

```python
from sklearn.preprocessing import OneHotEncoder

categorical_features = ['Gender']

categorical_features=[col for col in categorical_features if col in data.columns]

# Aplicar OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
encoded_categories = encoder.fit_transform(data[categorical_features])


columnas_codificadas=encoder.get_feature_names_out(['Gender'])

datos_codificados=pd.DataFrame(encoded_categories,columns=columnas_codificadas)
#print(datos_codificados)
# Concatenar el DataFrame codificado con el original
data2 = pd.concat([data.drop(columns=categorical_features), datos_codificados], axis=1)
data2
```

### Escalado de caracteristicas numericas
#### Uso: Escalar las caracteristicas numericas para que tengan una media   0 y una desviacion estandar de 1.
#### Razon de Uso: Muchos algoritmos de aprendizaje automático, especialmente aquellos que se basan en la distancia (como K-means y SVM), funcionan mejor cuando las características numéricas están en la misma escala.

```python
from sklearn.preprocessing import StandardScaler

numerical_features=['Total','Rating']
scaler=StandardScaler()

data[numerical_features]=scaler.fit_transform(data[numerical_features])
data
```

### Balanceo de clases (submuestreo)
#### Uso: Reducir el desequilibrio de clases en el dataset.
#### Razón de Uso: El desequilibrio de clases puede causar que los modelos favorezcan la clase mayoritaria. Submuestrear la clase mayoritaria ayuda a equilibrar las clases.

```python
from sklearn.utils import resample
data_majority=data[data['Gender']=='Female']
data_minority=data[data['Gender']=='Male']

data_majority_downsampled=resample(data_majority,replace=False,n_samples=len(data_minority),random_state=42)

data_balanced=pd.concat([data_minority,data_majority_downsampled])
data_balanced
```

En el contexto del conjunto de datos "bank marketing", la imputación puede ser necesaria debido a la presencia de valores faltantes en las características o atributos de los registros. Estos valores faltantes pueden surgir por diversas razones, como errores en la recopilación de datos, información incompleta o simplemente porque algunos registros no proporcionan ciertos detalles.

La imputación se utiliza para llenar estos valores faltantes de manera que el conjunto de datos sea más completo y adecuado para su análisis.
