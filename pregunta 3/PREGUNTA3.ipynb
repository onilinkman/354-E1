**3. Del dataset elegido realice la imputación por columnas sin scikit-learn.**

```python
import pandas as pd
dat = pd.read_csv("./bank.csv")
dat.head(10)
```

```python
import pandas as pd
pd.options.mode.chained_assignment = None
imp = dat.iloc[:, [0, 5, 9, 11, 12, 13, 14]]
# Imputar valores faltantes en las columnas seleccionadas con la mediana de cada columna
for column in imp.columns:
    median = imp[column].median()
    imp[column].fillna(median, inplace=True)
imp.head(10)
```

En el contexto del conjunto de datos "bank marketing", la imputación puede ser necesaria debido a la presencia de valores faltantes en las características o atributos de los registros. Estos valores faltantes pueden surgir por diversas razones, como errores en la recopilación de datos, información incompleta o simplemente porque algunos registros no proporcionan ciertos detalles.

La imputación se utiliza para llenar estos valores faltantes de manera que el conjunto de datos sea más completo y adecuado para su análisis.
