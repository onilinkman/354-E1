```python
pd.set_option('display.max_rows', None)  # Mostrar todas las filas
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
data=pd.read_csv("work/supermarket_sales - Sheet1.csv", header= 0, encoding= 'unicode_escape')
data.head()
```

## Inciso a) Con Python sin uso de librerías, calcule del ultimo cuartil, percentil 80 por columna; explique qué significa en cada caso.

### paso 1 ordenamos

```python
unit_prices=data["Unit price"]
sort_unit_prices=sorted(unit_prices)
quantity=data["Quantity"]
sort_quantity=sorted(quantity)
tax=data["Tax 5%"]
sort_tax=sorted(tax)
percentage=data["gross margin percentage"]
sort_percentage=sorted(percentage)
gross_income=data["gross income"]
sort_gross_income=sorted(gross_income)
rating=data["Rating"]
sort_rating=sorted(rating)
```

### paso 2 Usando la formula Q_k =(k*n)/4

```python
print("25% para unit_prices")
q1_unitPrices=(len(sort_unit_prices))/4
print("Q1 25%:",(sort_unit_prices[int(q1_unitPrices)-1]+sort_unit_prices[int(q1_unitPrices)])/2)
q2_unitPrices=(2*len(sort_unit_prices))/4
print("Q2 50%:",(sort_unit_prices[int(q2_unitPrices)-1]+sort_unit_prices[int(q2_unitPrices)])/2)
q3_unitPrices=(3*len(sort_unit_prices))/4
print("Q3 50%:",(sort_unit_prices[int(q3_unitPrices)-1]+sort_unit_prices[int(q3_unitPrices)])/2)
```

```python
print("para Quantity")
q1_quantity=(len(sort_quantity))/4
print("Q1 25%:",(sort_quantity[int(q1_quantity)-1]+sort_quantity[int(q1_quantity)])/2)
q2_quantity=(2*len(sort_quantity))/4
print("Q1 50%:",(sort_quantity[int(q2_quantity)-1]+sort_quantity[int(q2_quantity)])/2)
q3_quantity=(3*len(sort_quantity))/4
print("Q1 50%:",(sort_quantity[int(q3_quantity)-1]+sort_quantity[int(q3_quantity)])/2)
```

```python
print("para Tax 5%")
q1_tax=(len(sort_tax))/4
print("Q1 25%:",(sort_tax[int(q1_tax)-1]+sort_tax[int(q1_tax)])/2)
q2_quantity=(2*len(sort_tax))/4
print("Q1 50%:",(sort_tax[int(q2_quantity)-1]+sort_tax[int(q2_quantity)])/2)
q3_quantity=(3*len(sort_tax))/4
print("Q1 50%:",(sort_tax[int(q3_quantity)-1]+sort_tax[int(q3_quantity)])/2)
```

```python
print("para procentage")
q1_percentage=(len(sort_percentage))/4
print("Q1 25%:",(sort_percentage[int(q1_percentage)-1]+sort_percentage[int(q1_percentage)])/2)
q2_percentage=(2*len(sort_percentage))/4
print("Q1 50%:",(sort_percentage[int(q2_percentage)-1]+sort_percentage[int(q2_percentage)])/2)
q3_percentage=(3*len(sort_percentage))/4
print("Q1 50%:",(sort_percentage[int(q3_percentage)-1]+sort_percentage[int(q3_percentage)])/2)
```

```python
print("para gross income")
q1_gross=(len(sort_gross_income))/4
print("Q1 25%:",(sort_gross_income[int(q1_gross)-1]+sort_gross_income[int(q1_gross)])/2)
q2_gross=(2*len(sort_gross_income))/4
print("Q1 50%:",(sort_gross_income[int(q2_gross)-1]+sort_gross_income[int(q2_gross)])/2)
q3_gross=(3*len(sort_gross_income))/4
print("Q1 50%:",(sort_gross_income[int(q3_gross)-1]+sort_gross_income[int(q3_gross)])/2)
```

```python
print("para Rating")
q1_rating=(len(sort_rating))/4
print("Q1 25%:",(sort_rating[int(q1_rating)-1]+sort_rating[int(q1_rating)])/2)
q2_rating=(2*len(sort_rating))/4
print("Q1 50%:",(sort_rating[int(q2_rating)-1]+sort_rating[int(q2_rating)])/2)
q3_rating=(3*len(sort_rating))/4
print("Q1 50%:",(sort_rating[int(q3_rating)-1]+sort_rating[int(q3_rating)])/2)
```

```python
#Percentil 80
#con la formula p=(p/100)*(n+1)
p=(80/100)*(1000+1)
valor1=sort_unit_prices[int(p)-1]
valor2=sort_unit_prices[int(p)]
percentil_80=valor1+(p-int(p))*(valor2-valor1)
print("El percentil 80 de unit_prices es:",percentil_80)

p=(80/100)*(1000+1)
valor1=sort_quantity[int(p)-1]
valor2=sort_quantity[int(p)]
percentil_80=valor1+(p-int(p))*(valor2-valor1)
print("El percentil 80 de quantity es:",percentil_80)

p=(80/100)*(1000+1)
valor1=sort_tax[int(p)-1]
valor2=sort_tax[int(p)]
percentil_80=valor1+(p-int(p))*(valor2-valor1)
print("El percentil 80 de tax es:",percentil_80)

p=(80/100)*(1000+1)
valor1=sort_percentage[int(p)-1]
valor2=sort_percentage[int(p)]
percentil_80=valor1+(p-int(p))*(valor2-valor1)
print("El percentil 80 de percentage es:",percentil_80)

p=(80/100)*(1000+1)
valor1=sort_rating[int(p)-1]
valor2=sort_rating[int(p)]
percentil_80=valor1+(p-int(p))*(valor2-valor1)
print("El percentil 80 de rating es:",percentil_80)

p=(80/100)*(1000+1)
valor1=sort_gross_income[int(p)-1]
valor2=sort_gross_income[int(p)]
percentil_80=valor1+(p-int(p))*(valor2-valor1)
print("El percentil 80 de gross income es:",percentil_80)
```
